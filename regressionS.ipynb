```python
from statistics import mean 
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.array([1,2,3,4,5,6], dtype=np.float64)
ys = np.array([5,4,6,5,6,7], dtype=np.float64)
```

```python
def slope(x_values, y_values):
    m = ( 
        
        ( (mean(x_values)*mean(y_values)) - mean(x_values*y_values) ) 
        / 
        ((mean(x_values)**2) - mean(x_values**2) ) 
        
        )
    b = mean(y_values) - (m*mean(x_values))
    
    return m, b

m, b  = slope(xs,ys)
```

```python
regression_line = [(m*x)+b for x in xs]
```

```python
def squared_error(y_org, y_line):
    return sum((y_line - y_org)**2)

#coefficient of determination
def coefficient_of_dermination(y_original, y_line):
    print(y_original)
    print(y_line)
    y_mean_line =  [mean(y_original) for y in y_original]
    squared_error_regression =  squared_error(y_original, y_line)
    squared_error_y_mean = squared_error(y_original, y_mean_line)
    return (1-(squared_error_regression/squared_error_y_mean))
```

```python
plt.scatter(xs,ys)
plt.plot(xs, [mean(ys) for i in ys], c='yellow') #y mean line
plt.plot(xs, regression_line, c='red')
plt.scatter(5, 9, c='green')
```

```python
print(coefficient_of_dermination(ys, regression_line))
```

```python
import random

def create_dataset(hm, variance, step=2, correlation=False):
    val =1
    y_values = []
    for i in range(hm):
        y =  val + random.randrange(-variance, variance)
        y_values.append(y)
        if correlation and correlation == 'pos':
            val += step
        elif correlation and correlation == 'neg':
            val -= step

    x_values =  [i for i in range(len(y_values))]
    return np.array(x_values, dtype=np.float64), np.array(y_values, dtype=np.float64)
```

```python
x1, y1 =  create_dataset(40, 10, 2, correlation='pos')
m1, b1  = slope(x1,y1)
regression_line1 =  [(m1*x)+b1 for x in x1]
```

```python
plt.scatter(x1,y1)
plt.plot(x1, regression_line1)
```
